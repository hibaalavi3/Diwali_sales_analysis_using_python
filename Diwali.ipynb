```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("Diwali sales data.csv", encoding= 'unicode_escape')
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df=df.drop(["Status", "unnamed1"], axis=1)
df.head()
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
df.head()
```

```python
df.isnull().sum()
```

```python
df['Amount'] = df['Amount'].astype('int')
df["Amount"].dtypes
```

```python
df.describe()
```

```python
df[['Age', 'Orders', 'Amount']].describe()
```

```python
df.Gender.value_counts()
```

### Exploratory Data Analysis

```python
ax=sns.countplot(x='Gender', data=df, hue = 'Gender', legend=False)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen, palette="deep")
plt.show()
```

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age, hue='Age Group')
plt.show()
```

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sales_state
```

```python
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x="State", data=sales_state, y="Orders", hue="State")
plt.show()
```

```python
sale_state=df.groupby("State", as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False).head()
sale_state
```

```python
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x="State", y="Amount", data=sale_state, hue="State")
plt.show()
```

```python
mar=df.groupby(["Marital_Status", "Gender"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)
mar
```

```python
sns.set(rc={'figure.figsize':(6,5)})
ax=sns.barplot(x="Marital_Status", data=mar, hue="Gender", y="Amount")
```

```python
c=df["Marital_Status"].value_counts()
c
```

```python
ax=sns.countplot(x="Marital_Status", data=df, hue="Marital_Status")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sns.set(rc={'figure.figsize':(18,5)})
ax=sns.countplot(x="Occupation", data=df, hue="Occupation", palette="deep")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount', hue="Occupation", palette="deep")
plt.show()
```

```python
sns.set(rc={'figure.figsize':(20,8)})
ax = sns.countplot(data = df, x = 'Product_Category', hue="Product_Category", palette="deep")

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount', hue="Product_Category")
plt.show()
```

```python
df.groupby("Product_ID")["Orders"].sum().nlargest(10).sort_values(ascending=False).plot(kind="bar")
plt.show()
```
